### 2. subplots / twinx

* twinx

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
fig, ax = plt.subplots()

ax2 = ax.twinx()
ax.plot(x1, y1, c='r', label = 'data1')
ax2.plot(x2, y2, c='b', label = 'data2')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y1', fontsize = 15)
ax2.set_ylabel('y2', fontsize = 15)
```

* 한번에 그리는 경우 범례 표시
  * 두 차트를 합쳐서 legend에 넣어줘야함

```python
fig, ax = plt.subplots()

ax2=ax.twinx()
c1 = ax.plot(x1, y1, c='r', label='data1')
c2 = ax2.plot(x2,y2,c='b',label='data2')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y1', fontsize = 15)
ax2.set_ylabel('y2', fontsize = 15)

c = c1+c2
ax.legend(c, ['data1','data2'])
```

### 1. scatter : 좌표상에 데이터의 위치(분포) 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('iris')
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data=df, x = 'sepal_width', y = 'sepal_length')
```

```python
sns.scatterplot(data=df, x = 'sepal_width', y = 'sepal_length', hue='species')
```

### 2. scatterplot으로 특정 기준을 설정하여 분류

* petal_length 의 값을 3 미만, 5 미만, 5 이상으로 분류

```python
def clf(x):
    if x<3:
        return 'a'
    elif x<5:
        return 'b'
    else:
        return 'c'
```

```python
df['petal_length_clf'] = df['petal_length'].apply(clf)
df.head()
```

* style만 적용

```python
sns.scatterplot(data=df, x = 'sepal_width', y = 'sepal_length', style='petal_length_clf')
```

* 기존 분류를 hue로 설정한 기준을 style로 지정

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', style = 'petal_length_clf')
```

* 기존 분류를 hue로 설정한 기준을 size로 지정

```python
def size(x):
    if x < 3:
        return 3
    elif x < 5:
        return 5
    else:
        return 7
```

```python
df['petal_length_size'] = df['petal_length'].apply(size)
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', size = 'petal_length_size')
```

### 3. replot : 카테고리별로 따로 그리기

```python
sns.relplot(data = df, x = 'sepal_width', y = 'sepal_length', col = 'species')
```

### 4. lmplot : 회귀선을 같이 그림

```python
sns.lmplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

## [ heatmap ]

### 1. heatmap : 숫자로 표현된 것을 음양으로 표현 => pivot_table

* heatmap을 위해 pivot table을 생성

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('flights')
df.head()
```

```python
df.isnull().sum()
```

* 특정 기준에 따라 분류(피벗 테이블 형식) : 각 년도별 월별 승객수를 히트맵으로

```python
pivot_data = df.pivot(values = 'passengers', index = 'month', columns='year')
pivot_data
```

```python
sns.heatmap(data=pivot_data)
```

* 구분선 추가

```python
sns.heatmap(pivot_data, linewidths=.5)
```

* 색상 변경

```python
sns.heatmap(pivot_data, cmap="Blues")
```

* 해당 값 입력
  * 보기 불편한 경우 fmt로 형태 변경 가능

```python
sns.heatmap(pivot_data, cmap="Blues", annot = True)
```

```python
# 정수 형태로 입력

sns.heatmap(pivot_data, cmap="Blues", annot = True, fmt = 'd')
```

### 2. colorbar도 같이 그리기 : 색상이 의미하는 바를 표현

```python
fig, (_ax, _cbar_ax) = plt.subplots(2)
```

* cbar 수평(가로)로 표현

```python
sns.heatmap(pivot_data, ax=_ax,
                  cbar_ax=_cbar_ax,
                  cbar_kws={"orientation": "horizontal"})
fig
```

```python
grid_kws = {"height_ratios": (.85, .10), "hspace": 0.4} # cbar 크기 조정(85:10, 간격 0.4)

f, (_ax, _cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
sns.heatmap(pivot_data, ax=_ax,
                 cbar_ax=_cbar_ax,
                 cbar_kws={"orientation": "horizontal"})
```

* cbar 수직(세로)로 표현

```python
fig, (_ax, _cbar_ax) = plt.subplots(2)
```

```python
sns.heatmap(pivot_data, ax=_ax,
                  cbar_ax=_cbar_ax,
                  cbar_kws={"orientation": "vertical"})
fig
```

* 전체 수식의 결과
  * 비율 조정, cbar, 수평/수직, 간격, 색상, 수치입력

```python
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (a, c) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=a,
                 cbar_ax=c,
                 cbar_kws={"orientation": "horizontal"},
                 cmap="Blues", 
                 annot = True, fmt = 'd')
```
