## [ 단일 차트 ]

### 1. plot 조정

* plot : 직선

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [4, 5, 6]

plt.figure(figsize=(20,8))
plt.plot(x, y)
```

* x 값이 없으면 순서대로 0부터 지정됨

```python
plt.plot(y)
```

* lindwidth : 선 굵기 조정

```python
plt.plot(x, y, linewidth=10)
```

* color : 색상 조정

```python
plt.plot(x, y, color='r')
```

* marker : 데이터 위치 표시
  * 참고 : [marker 형태](https://matplotlib.org/stable/api/markers_api.html)

```python
plt.plot(x, y, marker='^')
```

```python
plt.plot(x, y, marker='o', markersize=10)
```

* linestyle : 선의 형태

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, 'ro')
```

### 2. 그래프 이름 및 축 조정

* title : 이름/크기 지정

```python
plt.plot(x, y)
plt.title('title', fontsize=20)
```

* 축 명

```python
plt.plot(x, y)
plt.xlabel('x_axis', fontsize=15)
plt.ylabel('y_axis')
```

* 그리드 설정

```python
plt.plot(x, y)
plt.grid(True)
```

## [ 여러 차트 하나에 그리기 ]

### 1. plt.plot 반복

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 4, 7]
```

```python
plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
```

```python
plt.plot(x1, y1, 'b', x2, y2, 'r')
```

* 범례 설정 및 위치 설정

```python
plt.plot(x1, y1, color = 'blue', label = 'data1')
plt.plot(x2, y2, color = 'red', label = 'data2')
plt.legend(loc='upper right')
```

```python
plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
plt.legend(['data1', 'data2'], loc = 'upper right', fontsize = 20)
```

### 2. subplots / twinx

* twinx

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
fig, ax = plt.subplots()

ax2 = ax.twinx()
ax.plot(x1, y1, c='r', label = 'data1')
ax2.plot(x2, y2, c='b', label = 'data2')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y1', fontsize = 15)
ax2.set_ylabel('y2', fontsize = 15)
```

* 한번에 그리는 경우 범례 표시
  * 두 차트를 합쳐서 legend에 넣어줘야함

```python
fig, ax = plt.subplots()

ax2=ax.twinx()
c1 = ax.plot(x1, y1, c='r', label='data1')
c2 = ax2.plot(x2,y2,c='b',label='data2')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y1', fontsize = 15)
ax2.set_ylabel('y2', fontsize = 15)

c = c1+c2
ax.legend(c, ['data1','data2'])
```

## [ 여러 차트 나눠 그리기 ]

### 1. subplot(a,b,c) : a행  b열의 c번째 plot 그리기

* subplot(행의 수, 열의 수, 해당 그래프가 그려질 위치)

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

```python
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.title('data2')
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')
plt.tight_layout() # 레이아웃 자동 설정
```

### 2. subplots(nrows, ncols)

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x1, y1, color = 'blue')
ax[1].plot(x2, y2, color = 'red')
fig
```
